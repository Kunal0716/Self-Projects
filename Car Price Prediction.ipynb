```python
df2 = df.select_dtypes(exclude = 'object')
df2
```

```python
for i in df2.columns:
    plt.boxplot(df2[i])
    
    plt.show()
```

```python
for i in df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven', 'Owner']]:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3 - Q1
    LL = Q1 - 1.5 * IQR 
    UL = Q3 + 1.5 * IQR 
    df = df[(df[i] >= LL) & (df[i] <= UL)]
    
df
```

```python
category_map1 = {'Petrol':0, 'Diesel':1, 'CNG':2}
category_map2 = {'Dealer':0, 'Individual':1}
category_map3 = {'Manual':0, 'Automatic':1}

df['Fuel_Type'] = df['Fuel_Type'].map(category_map1)
df['Seller_Type'] = df['Seller_Type'].map(category_map2)
df['Transmission'] = df['Transmission'].map(category_map3)
```

```python
df
```

```python
x = df.drop(columns = ['Car_Name', 'Selling_Price'])
x
```

```python
y = df['Selling_Price']
y
```

```python
df3 = df2.drop(columns = 'Selling_Price')
df3
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data = pd.DataFrame()
vif_data['Feature'] = df3.columns
vif_data['VIF'] = [variance_inflation_factor(df3.values, i) for i in range(len(df3.columns))]
vif_data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)
```

```python
x_train.shape, x_test.shape
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
perd = pd.DataFrame()
perd['Actual'] = y_test
perd['Predicted'] = y_pred
perd
```

```python
from sklearn import metrics

MAE = metrics.mean_absolute_error(y_test, y_pred)
MSE = metrics.mean_squared_error(y_test,y_pred)
RMSE = np.sqrt(MSE)
R2 = metrics.r2_score(y_test, y_pred)

print('MAE:', MAE)
print('MSE:', MSE)
print('RMSE:', RMSE)
print('R2:', R2)
```
